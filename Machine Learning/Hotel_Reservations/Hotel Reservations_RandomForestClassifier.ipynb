```python
from lazypredict.Supervised import LazyClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("Hotel Reservations.csv")
```

```python
df
```

```python
df.info()
```

### Data Preprocessing

```python
df.isnull().sum()
```

```python
df["booking_status"] = df["booking_status"].map({"Not_Canceled": 0, "Canceled": 1})
```

```python
df
```

```python
df=df.drop(['Booking_ID'],axis=1)
```

```python
df
```

```python
df.columns
```

```python
most_frequent_value=df['type_of_meal_plan'].mode()[0]
df['type_of_meal_plan'] = df['type_of_meal_plan'].replace('Not Selected', most_frequent_value)
df
```

```python
df = pd.get_dummies(df, columns=['type_of_meal_plan'])
df
```

```python
df = pd.get_dummies(df, columns=['room_type_reserved'])
df
```

```python
df = pd.get_dummies(df, columns=['market_segment_type'])
df
```

```python
y=df['booking_status']
X=df.drop(['booking_status'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state =123)
```

### Lazy Prediction to find the optimal algorithm

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

```python
#Random Forest Classifier is the optimal classifier for this data 
```

### Random Forest Classifier

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
